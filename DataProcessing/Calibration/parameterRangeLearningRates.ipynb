```python
rounds = 90000
maxEpochsLRTime = 40
preLearningExp = 128
decay1 = 0.99986
decay2 = 0.9999

epochsRounds = (rounds / preLearningExp) * maxEpochsLRTime

decayVals1 = [decay1]
decayVals2 = [decay2]
for i in range(0, (int)(epochsRounds)-1):
    decayVals1.append(decayVals1[i]*decay1)
    decayVals2.append(decayVals2[i]*decay2)


import matplotlib.pyplot as plt

# Plot Cournot values
plt.figure(figsize=(10, 6))
plt.plot(range((int)(epochsRounds)), decayVals1, label='Decay factor')
plt.plot(range((int)(epochsRounds)), decayVals2, label='Decay factor')
plt.title('Decay Over Time')
plt.xlabel('Time')
plt.ylabel('Decay')
plt.grid(True)
plt.legend()
plt.show()
```
